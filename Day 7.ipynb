```python
s="Good Morning"
print(s.upper())
```

```python
s="GOOD MORNING"
print(s.lower())
```

```python
s="GOOD MORNING"
print(s.islower())
```

```python
s="GOOD MORNING"
print(s.isupper())
```

```python
s="GOOD MORNING"
print(s.isalpha())
```

```python
s="GOOD MORNING"
print(s.find('G'))
```

```python
s="GOOD MORNING" #-1 indicates FALSE or not present
print(s.find('g'))
```

```python
s="GOOD MORNING"
print(s.count('G'))
print(s.count('g'))
print(s.count('a'))
print(s.count(' '))
```

```python
#Indexing
s="GOOD MORNING"
print(s[0])
print(s[5])
print(s[-2])
print(s[-7])
```

```python
s="helloworld!"
for i in range(len(s)):
  print(s[i])
```

```python
s="helloworld!"
for i in s:
  print(i)
```

```python
s="helloworld!"
for i in "good morning":
  print(i)
```

```python
#slicing
s="helloworld!"
print(s[1:5])
print(s[5:10])
print(s[0:8])
print(s[:5])
print(s[6:])
print(s[1:11:2])
```

```python
s="helloworld!"
print(s[10:5:-1])#negative indexing
print(s[-2:-10:-2])
```

```python
#palindrome
s=str(input("Enter a string "))
s=s.lower() #in order to make it non case sensitive
s=s.replace(" ","") #in order to remove space and compare (was it a bat i saw, a toyota a toyota)
#print(type(s))
reversed_s=s[::-1]
#print(reversed_s)
if s==reversed_s:
  print("PALINDROME")
else:
  print("Not a palindrome")
```
